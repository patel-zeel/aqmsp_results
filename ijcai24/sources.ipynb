```python
# acq_time has this format:
# 705 -> 07:05:00
# 435 -> 04:35:00
def get_time(x):
    x = str(x)
    if len(x) == 3:
        return "0" + x[0] + ":" + x[1:] + ":00"
    else:
        return x[0:2] + ":" + x[2:] + ":00"
df["date"] = pd.to_datetime(df["acq_date"] + " " + df["acq_time"].apply(lambda x: get_time(x)))
df.head(2)
```

```python
df['count'] = 1
# convert each time-stamp to its nearest hour
df["date"] = df["date"].dt.round("H")
df["date"] = df["date"] + pd.Timedelta('5 hours 30 minutes')
df.head(2)
```
