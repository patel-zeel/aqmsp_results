```python
import pandas as pd
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
with xr.open_dataset('/home/patel_zeel/aqmsp/aqmsp_data/datasets/cpcb/ijcai24/data.nc') as ds:
    pass
ds
```

```python
start = "2022-03-22"
end = "2022-04-08"
tmp_df = ds.sel(time=slice(start, end)).to_dataframe().reset_index()
tmp_df = tmp_df[tmp_df["CO_missing"] == False]
tmp_df = tmp_df[tmp_df["NO2_missing"] == False]
tmp_df = tmp_df.groupby("station").mean().sort_values("NO2", ascending=False)

# make a joint bar chart for pop_3km and NO2 for each station
fig, ax1 = plt.subplots(figsize=(8, 3))
# ax2 = ax1.twinx()
# tmp_df["CO"].plot(kind='bar', ax=ax1, color='b', width=0.4, position=1)
tmp_df["NO2"].plot(kind='bar', ax=ax1, color='r', width=0.4, position=0)
ax1.legend(bbox_to_anchor=(1.2, 1.1))
# ax2.legend(bbox_to_anchor=(1.2, 0.9))
ax1.set_xticklabels(tmp_df.index, rotation=90)
plt.tight_layout()
plt.show()

delhi = gpd.read_file("/home/patel_zeel/aqmsp/aqmsp_data/datasets/shapefiles/delhi/Districts.shp")

fig, ax = plt.subplots(figsize=(8, 8))
delhi.plot(ax=ax, color='white', edgecolor='black')
for var in ["PM2.5", "CO", "NO2"]:
    if var == "CO":
        m = 20
    else:
        m = 1
    plt.scatter(tmp_df["lon"], tmp_df["lat"], s=tmp_df[var]*m, c=tmp_df[var], cmap='coolwarm')
    
    vivek = tmp_df.loc['Vivek_Vihar_Delhi_DPCC']
    plt.colorbar(cmap='coolwarm')
    # plot arrows showing wind direction using U10 and V10
    tmp_df["U10_MpS"] = tmp_df["WS"] * np.cos(np.deg2rad(tmp_df["WD"]))
    tmp_df["V10_MpS"] = tmp_df["WS"] * np.sin(np.deg2rad(tmp_df["WD"]))
    
    plt.quiver(tmp_df["lon"], tmp_df["lat"], tmp_df["U10_MpS"], tmp_df["V10_MpS"], scale=20)
    plt.scatter(vivek['lon'], vivek['lat'], s=20, marker='x', c='g', label="Vivek Vihar DPCC")
    plt.scatter(77.328109, 28.625020, c='k', s=20, label="Ghazipur landfill", marker='x')
    plt.legend()
    plt.title(f"{var}: from {start} to {end}")
    plt.show()
    
tmp_df[["lat", "lon", "PM2.5", "CO", "NO2", "U10_MpS", "V10_MpS"]].sort_values("PM2.5", ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(20, 4))
ax2 = ax.twinx()
# s = "Vivek_Vihar_Delhi_DPCC"
# s = "Anand_Vihar_Delhi_DPCC"
for s in ["Indirapuram_Ghaziabad_UPPCB", "Vivek_Vihar_Delhi_DPCC", "Anand_Vihar_Delhi_DPCC"][:1]:
# ds.sel(time=slice(start, end)).sel(station=s)["PM2.5"].plot(ax=ax, label="PM2.5", color='r')
# ds.sel(time=slice(start, end)).sel(station=s)["NO2"].plot(ax=ax, label="NO2", color='b')
    ds.sel(time=slice(start, end)).sel(station=s)["CO"].plot(ax=ax, label=f"CO: {s}", color="C0")
    ds.sel(time=slice(start, end)).sel(station=s)["NO2"].plot(ax=ax2, label=f"NO2: {s}", color="C1")
    ds.sel(time=slice(start, end)).sel(station=s)["PM2.5"].plot(ax=ax2, label=f"PM2.5: {s}", color="C2")

# put a vline at the time of the fire. put time in timestamp format
ax.axvline(pd.Timestamp("2022-03-27 20:00:00"), color='k', linestyle='--', label="fire start")
ax.axvline(pd.Timestamp("2022-03-31 00:00:00"), color='k', linestyle='--', label="fire ends")

# (met_and_aq_ds.sel(time=slice(start, end)).sel(station="Vivek_Vihar_Delhi_DPCC")["NO2_missing"]*50).plot(ax=ax, label="missing", color='r')
ax.set_ylabel("CO")
# color y-axis labels of ax2
for label in ax.get_yticklabels():
    label.set_color('C0')

for label in ax2.get_yticklabels():
    label.set_color('C1')
    
ax.set_ylabel("CO", color='C0')
ax2.set_ylabel("NO2", color='C1')
ax.legend(loc='upper left')
ax2.legend(loc='upper right');
plt.show()

gfas = xr.open_dataset("/tmp/adaptor.mars_constrained.internal-1708340212.0168858-31320-4-ea35c9aa-bef2-4792-92ae-b4b19421241e.nc")

lat = tmp_df.loc["Indirapuram_Ghaziabad_UPPCB"]["lat"]
lon = tmp_df.loc["Indirapuram_Ghaziabad_UPPCB"]["lon"]
gfas

fig, ax = plt.subplots(figsize=(8, 8))
gfas.sel(latitude=slice(lat+1.0, lat-1.0), longitude=slice(lon-1.0, lon+1.0))["cofire"].isel(time=11).plot(ax=ax)
delhi.plot(ax=ax, color='none', edgecolor='black')
```
